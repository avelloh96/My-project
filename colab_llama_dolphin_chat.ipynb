```python
# 3️⃣ Path model Dolphin
model_path = '/content/drive/MyDrive/LLM_MODEL/Dolphin3.0-Llama3.1-8B-Q4_K_M.gguf'
```

```python
# 4️⃣ Fungsi Chat Streaming
import subprocess

def chat_stream(prompt):
    cmd = [
        '/content/llama.cpp/build/bin/llama-cli',
        '-m', model_path,
        '--color',
        '--temp', '0.7',
        '-p', prompt
    ]
    process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)
    for line in iter(process.stdout.readline, ''):
        print(line, end='')
    process.stdout.close()
    process.wait()
```

```python
# 5️⃣ Tanya chatbot (ubah teks untuk pertanyaan baru)
chat_stream("Halo, siapa namamu?")
```
